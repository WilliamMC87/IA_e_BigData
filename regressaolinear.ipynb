```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
lojas = ['A', 'B', 'C', 'D']
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criar DataFrame
df_vendas = pd.DataFrame(vendas_2022, index=lojas)

# Criar uma figura com 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)

# Lista de lojas para iterar
for i, loja in enumerate(lojas):
    ax = axes[i // 2, i % 2]  # Determinar a posio do subplot
    ax.plot(df_vendas.columns, df_vendas.loc[loja], marker='o')
    ax.set_title(f'Loja {loja}')
    ax.set_xlabel('Ms')
    ax.set_ylabel('Nmero de Vendas')
    ax.grid(True)

# Ajustar layout
plt.tight_layout()
plt.show()
```
